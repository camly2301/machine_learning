```python

train_data = [(2, 3), (3, 4), (4, 2), (4, 4), (1, 2), (2, 1),(4, 3)]
train_labels = ['A', 'A', 'A', 'B', 'B', 'B', 'B']

test_point = (3, 3)
k = 3

def euclidean_distance(point1, point2):
    distance = ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5 
    return distance
 
distances = [] 
for i, point in enumerate(train_data):                  # enumerate cho phép lặp và điếm số lần lặp
    distance = euclidean_distance(point, test_point)   
    distances.append((distance, i))                    
distances.sort()                                       
neighbors = [train_labels[i] for _, i in distances[:k]] 

# biến "_" để loại bỏ distance trong distances[] chỉ lấy i 
# chọn ra K điểm láng giềng gần nhất bằng cách lấy k phần tử đầu tiên của danh sách distances và trích xuất nhãn lớp tương ứng của chúng từ danh sách train_labels.

counts = {}                                 # Đếm số lần xuất hiện của mỗi nhãn lớp trong danh sách neighbors
for label in neighbors:
    if label in counts:
        counts[label] += 1                  # nếu các nhẵn lớp đã có trong count thì ta tăng giá trị trong count lên 1
    else:
        counts[label] = 1                   # nếu không thì khởi tạo giá trị đó là 0 
prediction = max(counts, key=counts.get)    # hàm max để tìm nhãn lớp có giá trị lớn nhất trong từ điển count
#get lấy giá trị của phẩn tử trong counts
#key là cái đk để mình lấy max 


print(f"Nhãn dự đoán cho điểm kiểm tra là: Điểm {test_point} là {prediction}.")
   
```
